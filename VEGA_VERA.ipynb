```python
#Apartado 3.3
print("3.3: ¿Qué lugar ocupa España si los ordenamos los vértices según sus valores de autocentralidad?")
import networkx as nx
G = graphs.EuropeMap()
L = {i: G.neighbors(i) for i in G}
centro = nx.Graph(L)
listaVertices = centro.nodes()
listaAristas = centro.edges()
lista = []
X = nx.eigenvector_centrality(centro)
resultado = sorted(X.items(), key = lambda v: v[1])

maximo = max(X.items(), key = lambda v: v[1])
print("     El de mayor centralidad propia es " + str(maximo) + "\n")

print("     ¿Cuál es el país con mayor centralidad propia?")
for i in range(len(listaVertices)):
    if(resultado[i][0] == "Spain"):
        print("     Posición", str(i+1), ":",resultado[i])
```

```python
#Apartado 3.4
print("3.4: ¿Qué lugar ocupa España si los ordenamos los vértices según sus valores de PageRank, con  α=0.85?")
europa = graphs.EuropeMap()
grafopg4 = europa.pagerank(0.85)
listaVertices = europa.vertices()
res4 = sorted(grafopg4.items(), key = lambda v: v[1], reverse=True)

for i in range(len(listaVertices)):
    if(res4[i][0] == "Spain"):
        print("     Posición", str(i+1), ":",res4[i],"\n")
        
print("     ¿Cuál es el país con mayor PageRank para ese valor de  α?")
mayor = max(res4, key = lambda v: v[1])
print("     El país con mayor PageRank para ese valor es: " + str(mayor) + "\n")

print("     ¿Cuál es el país con menor PageRank para ese valor de  α?")
menor = min(res4, key = lambda v: v[1])
print("     El país con minimo PageRank para ese valor es: " + str(menor) + "\n")
```
